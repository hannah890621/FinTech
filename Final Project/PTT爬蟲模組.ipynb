# 匯入套件

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def main(pattern):
    url = "https://www.ptt.cc/bbs/Stock/index.html"
    count = 0  #計數器
    final_info = []
    
    for times in range(5):
        res = requests.get(url)
        soup = BeautifulSoup(res.text, 'html.parser')
        articles = soup.select('.title a')
        paging = soup.select('.btn-group.btn-group-paging a')
        next_url = "https://www.ptt.cc" + paging[1]['href']
        url = next_url
        
        for each_title in articles:
            
            if pattern in each_title.text:
                url2 = "https://www.ptt.cc" + each_title['href']
                res2 = requests.get(url2)
                soup = BeautifulSoup(res2.text, 'html.parser')
                header = soup.select('.article-meta-value')

                author = header[0].text
                board = header[1].text
                title = header[2].text
                date = header[3].text.split(' ')[1] + ' ' + header[3].text.split(' ')[2] + ' ' + header[3].text.split(' ')[4]
                time = header[3].text.split(' ')[3]

                main_container = soup.find(id='main-container')
                all_text = main_container.text
                pre_text = all_text.split('--')[0]
                texts = pre_text.split('\n')
                contents = texts[2:]
                content = '\n'.join(contents)

                info = {
                    'author' : author,
                    'board' : board,
                    'title' : title,
                    'date' : date,
                    'time' : time,
                    'content' : content
                }
                final_info.append(info)
                
                count += 1
    df = pd.DataFrame(final_info)
    return df
```

```python
main('台積電')
```
