```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.facebook.com/pg/Gooaye/posts/"
res = requests.get(url)
soup = BeautifulSoup(res.text , 'html.parser')
```

```python
content = soup.select('._5pbx')
```

```python
for t in content:
    print('[' , t.text , ']')
```
