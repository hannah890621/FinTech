# 匯入套件

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# 主幹

```python
def main(co_id , year):
    form_data = {'encodeURIComponent': '1',
    'step': '1',
    'firstin': '1',
    'off': '1',
    'queryName': 'co_id',
    'inpuType': 'co_id',
    'TYPEK': 'all',
    'co_id': co_id,
    'year': year}

    res = requests.post("https://mops.twse.com.tw/mops/web/ajax_t05st01" , data = form_data)
    soup = BeautifulSoup(res.text, 'lxml')
    info = soup.select('.hasBorder')[0]
    df = pd.read_html(info.prettify('utf-8') , encoding= 'utf-8')
    return df[0]
```

```python
main(2330,108)
```
