```python
# basic
import numpy as np
import pandas as pd

# get data
import pandas_datareader as pdr

# visual
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import mplfinance as mpf
%matplotlib inline
import seaborn as sns

#time
import datetime as datetime

#talib
import talib
```

```python
start = datetime.datetime(2018,4,1)
end = datetime.datetime(2019,4,1)
df_2330 = pdr.DataReader('2330.TW', 'yahoo', start=start , end = end)
```

```python
df_2330.rename(columns={'Turnover':'Volume'}, inplace = True) 
mc = mpf.make_marketcolors(up='r',down='g', inherit=True)
s  = mpf.make_mpf_style(marketcolors=mc  , gridaxis = 'both')
kwargs = dict(type='candle', mav=(10,30) , volume=True, figratio=(50,20) , style = s ,datetime_format='%Y-%m-%d') 
mpf.plot(df_2330, **kwargs)
```
